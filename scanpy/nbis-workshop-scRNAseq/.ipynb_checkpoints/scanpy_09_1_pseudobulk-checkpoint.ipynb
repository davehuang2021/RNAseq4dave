When cell identity clusters are well-defined, it can be advantageous to perform analyses at the pseudobulk level rather than at the single-cell level. Pseudobulking involves aggregating counts across cells of the same type within each sample, effectively creating sample-level gene expression profiles per cell type. This approach helps mitigate the effects of technical noise and dropouts common in single-cell data, enabling the detection of lowly expressed genes that might otherwise be missed.

当细胞身份簇明确定义时，在pseudobulk水平（pseudobulk level）而非单细胞水平进行分析可能更具优势。
拟扩增是指聚合每个样本中同类型细胞的计数，从而有效地创建每个细胞类型的样本水平基因表达谱。
这种方法有助于减轻单细胞数据中常见的技术噪声和数据丢失的影响，从而能够检测到原本可能被遗漏的低表达基因。


Moreover, conducting differential expression analysis (DEA) at the pseudobulk level, treating each biological sample as the unit of observation, is statistically more robust. 
Unlike single-cell DEA, which assumes cells are independent (an assumption that is violated when cells originate from the same individual), sample-level pseudobulk analysis avoids **inflation of p-values** by reducing the number of observations and by correctly modeling biological replication [SGK+21: Jordan W. Squair, Matthieu Gautier, Claudia Kathe, Mark A. Anderson, Nicholas D. James, Thomas H. Hutson, Rémi Hudelle, Taha Qaiser, Kaya J. E. Matson, Quentin Barraud, Ariel J. Levine, Gioele La Manno, Michael A. Skinnider, and Grégoire Courtine. Confronting false discoveries in single-cell differential expression. Nature Communications, September 2021.](http://dx.doi.org/10.1038/s41467-021-25960-2).

此外，在拟本体水平进行差异表达分析 (DEA)，将每个生物样本视为观察单位，在统计上更为稳健。
与假设细胞相互独立（当细胞来自同一个体时，这一假设并不成立）的单细胞 DEA 不同，样本水平的pseudobulk分析通过减少观察次数和正确模拟生物重复，避免了 p 值的膨胀[ SGK+21 ] 。

The resulting gene-level statistics from pseudobulk DEA can then be used as input for downstream enrichment analyses.

然后可以将pseudobulk DEA 得到的基因级统计数据用作下游富集分析的输入。

## Pseudobulking


The pseudo-bulk approach involves the following steps:
伪批量方法包括以下步骤：

- Subsetting the cell type of interest. 对感兴趣的细胞类型进行子集设置
- Extracting their **raw integer counts**. 提取它们的原始整数计数
- Summing their counts per gene into a single profile if they pass quality control. 如果通过质量控制，则将每个基因的计数汇总成一个图谱

Then, DEA can be performed if there are **at least two biological replicates per condition** (more replicates are recommended).

然后，如果每个条件至少有两个生物学重复（建议有更多重复），则可以执行 DEA。

Pseudobulking can easily be performed using the function decoupler.pp.pseudobulk(). In this example, the counts are just summed, though other modes such as the mean or any custom aggregation function are available. For more information, refer to the mode argument.
使用函数 decoupler.pp.pseudobulk() 可以轻松实现Pseudobulking。在本例中，计数只是**相加**，但也可以使用**其他模式，例如平均值或任何自定义聚合函数**。有关更多信息，请参阅 mode 参数。

```python
pdata = dc.pp.pseudobulk(
    adata=adata,
    sample_col='individual',
    groups_col='celltype',
    mode='sum',
)
```

A profile has been generated for each sample and cell type. Quality control metrics can be then visualized.
已为每个样本和细胞类型生成了分析图谱。质量控制指标可实现可视化。

```python
dc.pl.filter_samples(
    adata=pdata,
    groupby=['disease', 'individual', 'celltype'],
    min_cells=10,
    min_counts=1000,
    figsize=(5, 8),
)
```
